```python
#https://chldkato.tistory.com/17
from collections import deque

dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]

def bfs(rx, ry, bx, by, cnt):
    q.append([rx, ry, bx, by])
    c.append([rx, ry, bx, by])
    while q:
        qlen = len(q)
        while qlen:
            rx, ry, bx, by = q.popleft()
            if a[rx][ry] == 'O':
                print(cnt)
                return
            for i in range(4):
                nrx, nry, nbx, nby = rx, ry, bx, by
                while True:
                    nrx += dx[i]; nry += dy[i]
                    if a[nrx][nry] == 'O':
                        break
                    if a[nrx][nry] == '#':
                        nrx -= dx[i]; nry -= dy[i]
                        break
                while True:
                    nbx += dx[i]; nby += dy[i]
                    if a[nbx][nby] == 'O':
                        break
                    if a[nbx][nby] == '#':
                        nbx -= dx[i]; nby -= dy[i]
                        break

                if a[nbx][nby] == 'O':
                    continue
                if nrx == nbx and nry == nby:
                    if abs(rx - nrx) + abs(ry - nry) > abs(bx - nbx) + abs(by - nby):
                        nrx -= dx[i]; nry -= dy[i]
                    else:
                        nbx -= dx[i]; nby -= dy[i]

                if [nrx, nry, nbx, nby] not in c:
                    c.append([nrx, nry, nbx, nby])
                    q.append([nrx, nry, nbx, nby])
            qlen -= 1

        if cnt == 10:
            print(-1)
            return
        cnt += 1
    print(-1)
    return

n, m = map(int, input().split())
a = [[] for _ in range(n)]
c, q, cnt = [], deque(), 0

for i in range(n):
        a[i] = list(map(str, input()))
        for j in range(m):
            if a[i][j] == 'R':
                rx, ry = i, j
                a[i][j] = '.'
            elif a[i][j] == 'B':
                bx, by = i, j;
                a[i][j] = '.'

bfs(rx, ry, bx, by, cnt)
```
