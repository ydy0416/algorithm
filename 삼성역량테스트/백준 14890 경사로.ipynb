```python
def garo(graph,row):
    tmp=[1 for i in range(n)]
    for i in range(n-1):
        #양 옆 마이너스 절대값이 1보다 크면 성립x
        if abs(graph[row][i]-graph[row][i+1])>1:
#             print('절대값1보다 클 때의 행', row)
            return 0
        #평지이면 유효한 땅 +1
        if graph[row][i]== graph[row][i+1]:
            tmp[i+1]=tmp[i]+1
        #다음 열이 더 클때
        elif graph[row][i]<graph[row][i+1]:
            #유효값이 0보다 작으면 리턴
            if tmp[i]-l<0:
#                 print('유효값이 0보다 작을 때 행', row, i)
                return 0
            #0보다 크면 다음 열 1로 초기화
            else:
                tmp[i+1]=1
        elif graph[row][i]>graph[row][i+1]:
            #이전 값의 유효값이 음수이면 리턴
            if tmp[i]<0:
                return 0
            tmp[i+1]=1-l
            
    if tmp[n-1]>=0:
        return 1
    else :
        return 0

def sero(graph,col):
    tmp=[1 for i in range(n)]
    for i in range(n-1):
        #위 아래 마이너스 절대값이 1보다 크면 성립x
        if abs(graph[i][col]-graph[i+1][col])>1:
#             print('절대값 1보다 클 때의 열', col)
            return 0
        #평지이면 유효한 땅+1
        if graph[i][col]==graph[i+1][col]:
            tmp[i+1]=tmp[i]+1
        #다음행이 더 클 때
        elif graph[i][col]<graph[i+1][col]:
            #유효값이 0보다 작으면 리턴
            if tmp[i]-l<0:
#                 print('유효값이 0보다 작을 때 열 ', col, i)
                return 0
            #0보다 크면 다음 유효값 1로 초기화
            else :
                tmp[i+1]=1
        elif graph[i][col]>graph[i+1][col]:
            #이전 값의 유효값이 음수이면 리턴
            if tmp[i]<0:
                return 0
            tmp[i+1]=1-l
    if tmp[n-1]>=0:
        return 1
    else :
        return 0
        
        
n,l=map(int,input().split())
graph=[]
for i in range(n):
    graph.append(list(map(int,input().split())))
# n,l=6,2
# graph=[
#  [3, 3, 3, 3, 3, 3],
#  [2, 3, 3, 3, 3, 3],
#  [2, 2, 2, 3, 2, 3],
#  [1, 1, 1, 2, 2, 2],
#  [1, 1, 1, 3, 3, 1],
#  [1, 1, 2, 3, 3, 2]]

ans=0
for i in range(n):
    ans+=garo(graph,i)
    ans+=sero(graph,i)
print(ans)
```
