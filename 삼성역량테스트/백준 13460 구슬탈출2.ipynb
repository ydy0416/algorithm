```python
from collections import deque

#동서남북 bfs
def bfs(queue,queue2,graph,visited):
    cnt=0
    while queue:
        x,y=queue.popleft()
        cnt+=1
        if graph[x][y]=='O':
            return cnt
        #동서남북 이동
        for i in range(4):
            #직진
            while True:
                nx=x+dx[i]
                ny=y+dy[i]
                print(nx,ny)
                #구멍 만나면
                if graph[nx][ny]=='O':
                    return cnt
                #범위이탈시 무시
                if nx<0 or nx>=n or ny<0 or ny>=m:
                    continue
                #빈칸 & 아직 방문 안한 곳
                if graph[nx][ny]=='.' and visited[nx][ny]==9999:
                    visited[nx][ny]=visited[x][y]
                #벽 만날시
                if graph[nx][ny]=='#':
                    break
        if i==3:
            break

#n,m 입력
n,m=map(int,input().split())
#그래프 입력받기
graph=[]
graph=[['#', '#', '#', '#', '#', '#', '#'],
 ['#', '.', '.', 'R', '#', 'B', '#'],
 ['#', '.', '#', '#', '#', '#', '#'],
 ['#', '.', '.', '.', '.', '.', '#'],
 ['#', '#', '#', '#', '#', '.', '#'],
 ['#', 'O', '.', '.', '.', '.', '#'],
 ['#', '#', '#', '#', '#', '#', '#']]
for i in range(n):
#     graph.append(list(input()))
    for j in range(m):
        if graph[i][j]=='R':
            #초기 빨간공의 위치
            rx,ry=i,j
        if graph[i][j]=='B':
            #초기 파란공의 위치
            bx,by=i,j

#동서남북 이동
dx=[0,0,1,-1]
dy=[1,-1,0,0]
#큐생성
queue=deque()
queue2=deque()
#빨간공,파란공 초기 큐
queue.append((rx,ry))
queue2.append((bx,by))

#방문기록
#빨간공
visited=[[9999]*(m) for i in range(n)]
#빨간공 자리 1로 초기화
visited[rx][ry]=1
#파란공
visited2=[[9999]*(m) for i in range(n)]

ans=bfs(queue,queue2,graph,visited)
print(ans)
```
