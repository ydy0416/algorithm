```python
# https://velog.io/@jwisgenius/%ED%94%84%EB%A1%9C%EA%B7%B8%EB%9E%98%EB%A8%B8%EC%8A%A4-LV2-%EB%A9%94%EB%89%B4-%EB%A6%AC%EB%89%B4%EC%96%BC

from itertools import combinations
orders=["ABCFG", "AC", "CDE", "ACDE", "BCFG", "ACDEH"]
course=[2,3,4]

orders2=[[] for i in range(len(orders))]
k=0
for i in orders:
    for j in i:
        orders2[k].append(j)
    k+=1

#중복없이 합친것
hap=[]
for i in orders:
    for j in i:
        hap.append(j)
hap=list(set(hap))
hap.sort()

#조합만들기
combi=[]
for i in course:
    case=list(combinations(hap,i))
    combi.append(case)



result=[]
for i in combi:
    check=0
    for j in i:
        print(j)
        if j in orders2:
            check+=1
        if check >=2:
            result.append(j)
```
