```python
#백준 15686번 치킨배달
from itertools import combinations
n,m=map(int,input().split())
graph=[]
chicken=[]
house=[]

#그래프 입력받기 & 치킨집 & 집 인덱스 추출
for i in range(n):
    graph.append(list(map(int, input().split())))
    for j in range(n):
        if graph[i][j]==2:
            chicken.append([i,j])
        if graph[i][j]==1:
            house.append([i,j])

#치킨집 조합
chicken_set=list(map(list,combinations(chicken,m)))

total=[]
# #집을 기준으로 치킨거리 총합구하기
# for h in house:
    
#     #치킨집 한세트
#     for c in chicken_set:
#         minimum=9999
#         ans=0 #치킨집 한 세트의 총합
        
#         #한 집과 가장 가까운 치킨집의 치킨거리구하기
#         for c_xy in c:
#             #치킨거리
#             distance=abs(h[0]-c_xy[0])+abs(h[1]-c_xy[1])
#             if distance<minimum:
#                 minimum=distance  #한 집에서 가장 가까운 치킨거리
#         #한 세트의 치킨거리 총합
#         ans+=minimum
#     #한세트총합을 추가
#     total.append(ans)

#치킨집 한세트
for c in chicken_set:
    ans=0
    #집 돌기
    for h in house:    
        minimum=9999
        
        #한 집과 가장 가까운 치킨집의 치킨거리구하기
        for c_xy in c:
            #치킨거리
            distance=abs(h[0]-c_xy[0])+abs(h[1]-c_xy[1])
            if distance<minimum:
                minimum=distance  #한 집에서 가장 가까운 치킨거리
        #한 세트의 치킨거리 총합
        ans+=minimum
    #한세트총합을 추가
    total.append(ans)
print(min(total))
#1. 치킨집, 집 인덱스받기
#2. 폐업시키지 않을 치킨집 조합 구하기
#3. 조합별로 모든 집과의 치킨거리 구하기
#4. 새로운 리스트를 만들어 모든집의 치킨거리의 합 추가하기
#5. 리스트 중 최솟값 출력
    
```
