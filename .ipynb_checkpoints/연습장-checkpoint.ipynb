```python
from collections import deque
def dfs(graph,cur,CCTV,check):
    global ans
    if cur==cnt:
        tmp=0
        for i in range(n):
            for j in range(m):
                if graph[i][j]==0 and not check[i][j]:
                    tmp+=1
        if tmp<ans:
            ans=tmp
        return
    #4방향 회전
    x,y,tv=CCTV[cur][0],CCTV[cur][1],CCTV[cur][2]
    print('x,y,tv:',x,y,tv)
    for i in range(4):
        new_dir=[]
        if tv==1:
            new_dir.append(i)
        elif tv==2:
            new_dir.append(i)
            new_dir.append((i+2)%4)
        elif tv==3:
            new_dir.append(i)
            new_dir.append((i+1)%4)
        elif tv==4:
            new_dir.append(i)
            new_dir.append((i+1)%4)
            new_dir.append((i+3)%4)
        q=deque()
        print('new_dir:', new_dir)
        for _ in new_dir:
            nx=x+dx[_]
            ny=y+dy[_]
            while 0 <= nx < n and 0 <= ny < m:
                if graph[nx][ny]==6:
                    break
                if check[nx][ny]==False:
                    check[nx][ny]=True
                    q.append([nx,ny])
                nx+=dx[_]
                ny+=dx[_]
    #그래프 출력
    for i in range(n):
        for j in range(m):
            print('%2s'%graph[i][j], end=' ')
        print()
    dfs(graph,cur+1,CCTV,check)
    while q:
        bx,by=q.popleft()
        check[bx][by]=False

n,m=map(int,input().split())
graph=[]
CCTV=[]
CCTV5=[]
cnt=0
check=[[False]*m for i in range(n)]
for i in range(n):
    graph.append(list(map(int,input().split())))
    for j in range(m):
        if 0<graph[i][j]<5:
            CCTV.append([i,j,graph[i][j]])
            cnt+=1
            check[i][j]=True
        elif graph[i][j]==5:
            CCTV5.append([i,j])
#시계방향
dx=[-1,0,1,0]
dy=[0,1,0,-1]
#CCTV5처리
if len(CCTV5)>0:
    for x,y in CCTV5:
        for i in range(4):
            nx=x+dx[i]
            ny=y+dy[i]
            while True:
                if nx<0 or nx>=n or ny<0 or ny>=m:
                    break
                graph[nx][ny]='#'
                nx+=dx[i]
                ny+=dy[i]
ans=2147000000
# dfs(graph, 0,CCTV,check)
print(CCTV)
```

```python
from itertools import combinations 
l,c=map(int,input().split())
data=list(input().split())
data1=[]
data2=[]
for i in data:
    if i=='a' or i=='e' or i=='i' or i=='o' or i=='u':
        data1.append(i)
    else:
        data2.append(i)
ans=[]
a=1
b=l-a
# print('data1, data2:', data1,data2)
while b>1:
    data3=list(combinations(data1,a))
    data4=list(combinations(data2,b))
#     print('data1, data2:', data1,data2)
    for i in data3:
        tmp=list(i)
        for k in data4:
            t=tmp+list(k)
            t.sort()
            ans.append(t)
    
    a+=1
    b=l-a
ans.sort()
for i in ans:
    print(''.join(i))
```

```python
def dfs(graph,v,visited,long,res):
    visited[v]=True
    #a는 연결된 노드, b는 거리
    for a,b in graph[v]:
        if not visited[a]:
            res[a]=long+b
            dfs(graph,a,visited,long+b,res)
    
res=[0]*(n+1)
visited=[False]*(n+1)
long=0
dfs(graph,1,visited,long,res)
# print(res)
maximum=0
for i in range(1,len(res)):
    if maximum<res[i]:
        maximum=res[i]
        index=i
# print(index)

#두번째 dfs
visited=[False]*(n+1)
long=0
dfs(graph,index,visited,long,res)
print(max(res))
```

```python
price=[34000,78000, 48000, 27000, 11000, 285000, 320000, 335100]
box=[]
#초기보유량, 잔고, 대출유무
for i in price:
    box.append([100000000//i,100000000%i,0])

ans=[]
for i in range(len(price)):    
    for j in range(i+1,len(price)):
#         print('box:',box[i])
#         print('산가격, 현재가격:', price[i],price[j])
#         print('총가치:',box[i][0]*price[j]+box[i][1])
        #10억달성
        if box[i][0]*price[j]+box[i][1]+box[i][2]>1000000000 :
#             print('달성시 현재와 산날:', j,i)
            ans.append(j-i)
            break
        #산것보다 50%쌀때
        if price[i]//2>=price[j]:
            #대출을 안받았을경우
            if box[i][2]==0:
                #보유수량
                box[i][0]+=(50000000+box[i][1])//price[j]
                #잔고
                box[i][1]=(50000000+box[i][1])%price[j]
                box[i][2]=-50000000
    if box[i][0]*price[j]+box[i][1]<1000000000:
        ans.append(-1)
```

```python
def dfs(queen, n, row):
    global count
    if n == row:
        count+=1
    
    # 가로로 한번만 진행
    for col in range(n):
        queen[row] = col

        # for-else구문
        for x in range(row):
            # 세로로 겹치면 안됨
            if queen[x] == queen[row]: 
                break
                
            # 대각선으로 겹치면 안됨
            if abs(queen[x]-queen[row]) == (row-x):
                break
        else:
            dfs(queen, n, row+1)
            


def solution(n):
    queen = [0]*n
    global count
    count=0
    return dfs(queen, n, 0)
solution(4)
print(count)
```
